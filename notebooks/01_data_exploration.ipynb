## 2. Load Dataset

```python
import pandas as pd

# Đường dẫn tới file CSV (đặt trong data/raw/)
CSV_PATH = '../data/raw/vietnamese_news.csv'

df = pd.read_csv(CSV_PATH, nrows=10_000)  # Load 10k dòng để thử
print(f'Shape: {df.shape}')
df.head()
```

```python
# Thông tin cơ bản
print('Columns:', df.columns.tolist())
print('\nMissing values:')
print(df.isnull().sum())
```

## 3. Phân phối Category

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 5))
cat_counts = df['category'].value_counts()
sns.barplot(x=cat_counts.index, y=cat_counts.values, palette='viridis')
plt.title('Phân phối Category')
plt.xlabel('Category')
plt.ylabel('Số bài viết')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
print(cat_counts)
```

## 4. Phân phối Độ dài Bài viết

```python
df['content_length'] = df['content'].fillna('').str.len()
print('Thống kê độ dài nội dung (ký tự):')
print(df['content_length'].describe())

plt.figure(figsize=(10, 4))
df['content_length'].clip(upper=5000).hist(bins=50, color='steelblue', edgecolor='white')
plt.title('Phân phối độ dài bài viết (clipped at 5000 chars)')
plt.xlabel('Số ký tự')
plt.ylabel('Số bài viết')
plt.tight_layout()
plt.show()
```

## 5. Thử nghiệm Tokenization tiếng Việt

```python
from underthesea import word_tokenize

sample_texts = [
    'học máy là gì',
    'du lịch Hà Nội mùa thu',
    'ứng dụng trí tuệ nhân tạo trong y tế',
]

for text in sample_texts:
    tokens = word_tokenize(text, format='text').split()
    print(f'Input  : {text}')
    print(f'Tokens : {tokens}')
    print()
```
